```python
import pandas as pd
import math
import pickle
import numpy as np
```

```python
a = pd.read_pickle("vectors.pickle")
vector = a.T
from matplotlib import pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
Error =[]
for i in range(1, 30):
    kmeans = KMeans(n_clusters = i).fit(vector.iloc[1:,:31983])
    kmeans.fit(vector.iloc[1:,:31983])
    Error.append(kmeans.inertia_)
    print(i)
import matplotlib.pyplot as plt
plt.plot(range(1, 30), Error)
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('Error')
plt.show()
```

```python
with open('cluster_errors.pickle','wb') as file:
    pickle.dump(Error,file)
```
