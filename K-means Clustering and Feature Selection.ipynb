```python
def calculatex2(x,dic):
    T = 31983
    x2 = [] #term -> class -> (x2score,pre_on,pre_off,ab_on,ab_off)
    x2class = []
    temp = list()
    for j in range(len(x)): #for each class
        x2class.append(temp.copy())

    for i in range(T): #for each term
        x2.append(temp.copy())
        index = dic.t_index[i]
        post = dic.posting[i]
        #print(index,post)
        for j in range(len(x)): #for each class
            x2[i].append(temp.copy())
            scores = []
            pre_on = 0
            pre_off = 0
            ab_on = 0
            ab_of = 0
            for k in range(len(x[j])): #for each doc
                if x[j][k] in post: #if apperance
                    pre_on += 1
            pre_off = len(post) - pre_on
            ab_on = len(x[j]) - pre_on
            ab_off = N - pre_on - pre_off - ab_on
            scores = [pre_on,pre_off,ab_on,ab_off]
            #print(scores)

            pre = pre_on + pre_off
            ab = ab_on + ab_off
            on = pre_on + ab_on
            off = pre_off + ab_off

            Epre_on = pre * on / N
            Epre_off = pre * off / N
            Eab_on = ab * on / N
            Eab_off = ab * off / N

            x2score = 0
            x2score += pow((pre_on - Epre_on),2)/Epre_on
            x2score += pow((pre_off - Epre_off),2)/Epre_off
            x2score += pow((ab_on - Eab_on),2)/Eab_on
            x2score += pow((ab_off - Eab_off),2)/Eab_off

            x2[i][j] = [x2score,pre_on,pre_off,ab_on,ab_off]
            x2class[j].append([index,x2score])
    return x2,x2class
```

```python
def sortclass(x2class,tokenN):
    c = []
    temp = []
    for i in range(len(x2class)):
        c.append(temp.copy())
        x2class[i] = sorted(x2class[i],key = lambda x : x[1],reverse = True)
    for i in range(len(x2class)):
        for j in range(tokenN):
            c[i].append(x2class[i][j])
    return c
```
